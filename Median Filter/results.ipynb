```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# ---------------- Upload Images ----------------
input_image_path = "Input/original.jpg"
pure_python_image_path = "Output/median_filtered_image_pure.jpg"
numpy_image_path = "Output/median_filtered_image_numpy.jpg"
cython_image_path = "Output/median_filtered_image_cython.jpg"

# Load grayscale images
original_image = cv2.imread(input_image_path, cv2.IMREAD_GRAYSCALE)
pure_python_image = cv2.imread(pure_python_image_path, cv2.IMREAD_GRAYSCALE)
numpy_image = cv2.imread(numpy_image_path, cv2.IMREAD_GRAYSCALE)
cython_image = cv2.imread(cython_image_path, cv2.IMREAD_GRAYSCALE)
```

```python
plt.figure(figsize=(6,6))
plt.imshow(original_image, cmap='gray')
plt.title("Original Image")
plt.axis("off")
plt.show()

plt.figure(figsize=(6,6))
plt.imshow(pure_python_image, cmap='gray')
plt.title("Pure Python Filter")
plt.axis("off")
plt.show()

plt.figure(figsize=(6,6))
plt.imshow(numpy_image, cmap='gray')
plt.title("NumPy filter")
plt.axis("off")
plt.show()

plt.figure(figsize=(6,6))
plt.imshow(cython_image, cmap='gray')
plt.title("Cython filter")
plt.axis("off")
plt.show()

```

```python
# ---------------- Load Runtimes ----------------
time_log_path = "TIMES"
execution_times = {}

with open(time_log_path, "r") as f:
    for line in f:
        method, time = line.strip().split(": ")
        execution_times[method] = float(time.split()[0])

# Create Time DataFrame
df_times = pd.DataFrame(list(execution_times.items()), columns=["Method", "Execution Time (s)"])
print(df_times)
```

```python
# ---------------- View Time Comparison ----------------
plt.figure(figsize=(8, 5))
plt.bar(df_times["Method"], df_times["Execution Time (s)"], color=['red', 'blue', 'green', 'purple'])
plt.xlabel("Method")
plt.ylabel("Time (s)")
plt.title("Execution Time Comparison")
plt.show()
```
